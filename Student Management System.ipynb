```python
from tkinter import *
from tkinter import ttk
import pymysql
from tkinter import messagebox
```

```python
class Student:
    
    def __init__(self, root):
        
        self.root = root
        self.root.title('Student Management System')
        self.root.geometry('1350x700+0+0')
        self.root.configure(bg = 'lightcoral')
        
        title = Label(self.root, text='Student Management System', bd=12, relief=GROOVE, bg='crimson', fg='white', font=('times new roman', 30, 'bold'), pady=2).pack(fill=X)
        
        self.roll_no = StringVar()
        self.name = StringVar()
        self.email = StringVar()
        self.gender = StringVar()
        self.contact = StringVar()
        self.dob = StringVar()
        
        self.search_by = StringVar()
        self.search_txt = StringVar()
        
        frame1 = Frame(self.root, bd=10, relief=GROOVE, bg='salmon')
        frame1.place(x=5, y=90, width=850, height=600)      
            
        f1_lb1 = Label(frame1, text='Search', bg='salmon', fg='white', font=('times new roman', 18, 'bold')).grid(row=0,column=0,padx=20,pady=15)
        f1_lb1_choice = ttk.Combobox(frame1, textvariable=self.search_by, width=10, font=('times new roman', 14), state='readonly')
        f1_lb1_choice['values'] = ('Name', 'Roll_No.', 'Email', 'Address')
        f1_lb1_choice.grid(row=0,column=1,padx=20,pady=15)
        
        f1_search = Entry(frame1, textvariable=self.search_txt, font=('times new roman', 14), bd=5, relief=RIDGE, width=20).grid(row=0,column=2,padx=18,pady=15,stick='w')
        
        searchbtn = Button(frame1, text='Search', command=self.search_data, width=8, bg='crimson', fg='white', font=('times new roman', 15, 'bold')).grid(row=0, column=3, padx=18, pady=15, stick='w')
        showallbtn = Button(frame1, text='Show All', command=self.fetch_data, width=8, bg='crimson', fg='white', font=('times new roman', 15, 'bold')).grid(row=0, column=4, padx=18, pady=15, stick='w')
        
        f1_table = Frame(frame1, bd=4, relief=RIDGE, bg='salmon')
        f1_table.place(x=25, y=80, width=760, height=480)
        
        f1_scroll_x = Scrollbar(f1_table, orient=HORIZONTAL)
        f1_scroll_x.pack(side=BOTTOM, fill=X)
        
        f1_scroll_y = Scrollbar(f1_table, orient=VERTICAL)
        f1_scroll_y.pack(side=RIGHT, fill=Y)
        
        self.st_table = ttk.Treeview(f1_table, columns=('Roll No.', 'Name', 'Email', 'Gender', 'Contact', 'DOB.', 'Address'), xscrollcommand=f1_scroll_x.set, yscrollcommand=f1_scroll_y.set)
        
        f1_scroll_x.config(command = self.st_table.xview)
        f1_scroll_y.config(command = self.st_table.yview)
        
        self.st_table.heading('Roll No.', text='Roll No.')
        self.st_table.heading('Name', text='Name')
        self.st_table.heading('Email', text='Email')
        self.st_table.heading('Gender', text='Gender')
        self.st_table.heading('Contact', text='Contact')
        self.st_table.heading('DOB.', text='DOB.')
        self.st_table.heading('Address', text='Address')
        
        self.st_table['show'] = 'headings'
        
        self.st_table.column('Roll No.', width=80)
        self.st_table.column('Name', width=100)
        self.st_table.column('Email', width=100)
        self.st_table.column('Gender', width=100)
        self.st_table.column('Contact', width=100)
        self.st_table.column('DOB.', width=100)
        self.st_table.column('Address', width=100)
        
        self.st_table.pack(fill=BOTH, expand=1)
        
        self.st_table.bind('<ButtonRelease-1>', self.get_clicked_data)
        self.fetch_data()
        
        frame2 = Frame(self.root, bd=10, relief=RIDGE, bg='salmon')
        frame2.place(x=895, y=90, width=450, height=60)
        
        f2_lb1 = Label(frame2, text='Student Details', bg='crimson', fg='white', font=('times new roman', 20, 'bold')).pack(fill=X)
        
        frame3 = Frame(self.root, bd=10, relief=RIDGE, bg='salmon')
        frame3.place(x=895, y=140, width=450, height=545)
        
        f3_lb1 = Label(frame3, text='Roll No.',bg='salmon', fg='white',font=('times new roman', 18, 'bold')).grid(row=0,column=0,padx=30,pady=20,sticky='w')
        f3_lb1_txt = Entry(frame3, textvariable=self.roll_no, font=('times new roman', 14), bd=5, relief=SUNKEN, width=20).grid(row=0,column=1,padx=30,pady=20)
        
        f3_lb2 = Label(frame3, text='Name',bg='salmon', fg='white',font=('times new roman', 18, 'bold')).grid(row=1,column=0,padx=30,pady=15,sticky='w')
        f3_lb2_txt = Entry(frame3, textvariable=self.name, font=('times new roman', 14), bd=5, relief=SUNKEN, width=20).grid(row=1,column=1,padx=30,pady=15)
        
        f3_lb3 = Label(frame3, text='Email',bg='salmon', fg='white',font=('times new roman', 18, 'bold')).grid(row=2,column=0,padx=30,pady=15,sticky='w')
        f3_lb3_txt = Entry(frame3, textvariable=self.email, font=('times new roman', 14), bd=5, relief=SUNKEN, width=20).grid(row=2,column=1,padx=30,pady=15)
        
        f3_lb4 = Label(frame3, text='Gender',bg='salmon', fg='white',font=('times new roman', 18, 'bold')).grid(row=3,column=0,padx=30,pady=15,sticky='w')
        f3_choice = ttk.Combobox(frame3, textvariable=self.gender, font=('times new roman', 14), state='readonly')
        f3_choice['values'] = ('Male', 'Female', 'Other')
        f3_choice.grid(row=3,column=1,padx=30,pady=15)
       
        f3_lb5 = Label(frame3, text='Contact',bg='salmon', fg='white',font=('times new roman', 18, 'bold')).grid(row=4,column=0,padx=30,pady=15,sticky='w')
        f3_lb5_txt = Entry(frame3, textvariable=self.contact, font=('times new roman', 14), bd=5, relief=SUNKEN, width=20).grid(row=4,column=1,padx=30,pady=15)

        f3_lb4 = Label(frame3, text='DOB.',bg='salmon', fg='white',font=('times new roman', 18, 'bold')).grid(row=5,column=0,padx=30,pady=15,sticky='w')
        f3_lb4_txt = Entry(frame3, textvariable=self.dob, font=('times new roman', 14), bd=5, relief=SUNKEN, width=20).grid(row=5,column=1,padx=30,pady=15)
        
        f3_lb4 = Label(frame3, text='Address',bg='salmon', fg='white',font=('times new roman', 18, 'bold')).grid(row=6,column=0,padx=30,pady=15,sticky='w')
        self.address = Text(frame3, width=23, height=1.5)
        self.address.grid(row=6,column=1,padx=30,pady=15)
        
        frame4 = Frame(frame3, bd=4, relief=GROOVE, bg='salmon')
        frame4.place(x=20, y=465, width=385)
        
        addbtn = Button(frame4, text='Add', command=self.add_student, width=7, bg='crimson', fg='white', font=('times new roman', 15, 'bold')).grid(row=0, column=0)
        updatebtn = Button(frame4, command=self.update_data, text='Update', width=7, bg='crimson', fg='white', font=('times new roman', 15, 'bold')).grid(row=0, column=1)
        deletebtn = Button(frame4, command=self.delete_data, text='Delete', width=7, bg='crimson', fg='white', font=('times new roman', 15, 'bold')).grid(row=0, column=2)
        clearbtn = Button(frame4, command=self.clear_data, text='Clear', width=7, bg='crimson', fg='white', font=('times new roman', 15, 'bold')).grid(row=0, column=3)
        
        
    def add_student(self):
        
        if self.roll_no.get()=='':
            messagebox.showerror('Error', 'All fields are required')
        else:
            con = pymysql.connect(host='localhost', user='root', password='', database='stm')
            cur = con.cursor()
            cur.execute('insert into students values(%s,%s,%s,%s,%s,%s,%s)', (self.roll_no.get(), 
                                                                             self.name.get(), 
                                                                             self.email.get(), 
                                                                             self.gender.get(), 
                                                                             self.contact.get(), 
                                                                             self.dob.get(), 
                                                                             self.address.get('1.0', END)
                                                                            ))

            con.commit()
            self.fetch_data()
            self.clear_data()
            con.close()
            messagebox.showinfo('Record Updated','Record updated successfully')
        
        
    def fetch_data(self):
        con = pymysql.connect(host='localhost', user='root', password='', database='stm')
        cur = con.cursor()
        cur.execute('select * from students')
        data = cur.fetchall()
        if len(data) != 0:
            self.st_table.delete(*self.st_table.get_children())
            for row in data:
                self.st_table.insert('', END, values=row)
            con.commit()
        con.close()
        
    def clear_data(self):
        self.roll_no.set('')
        self.name.set('')
        self.email.set('')
        self.gender.set('')
        self.contact.set('')
        self.dob.set('')
        self.address.delete('1.0', END)
        
    def get_clicked_data(self, event): 
        data = self.st_table.focus()
        contents = self.st_table.item(data)
        row = contents['values']
        
        self.roll_no.set(row[0])
        self.name.set(row[1])
        self.email.set(row[2])
        self.gender.set(row[3])
        self.contact.set(row[4])
        self.dob.set(row[5])
        self.address.delete('1.0', END)
        self.address.insert(END, row[6])
        
    def update_data(self):
        con = pymysql.connect(host='localhost', user='root', password='', database='stm')
        cur = con.cursor()
        cur.execute('update students set name=%s, email=%s, gender=%s, contact=%s, dob=%s, address=%s where roll_no=%s', ( 
            self.name.get(), 
            self.email.get(), 
            self.gender.get(), 
            self.contact.get(), 
            self.dob.get(), 
            self.address.get('1.0', END),
            self.roll_no.get()
        ))
        
        con.commit()
        self.fetch_data()
        self.clear_data()
        con.close()        
        
    def delete_data(self):
        con = pymysql.connect(host='localhost', user='root', password='', database='stm')
        cur = con.cursor()
        cur.execute('delete from students where roll_no=%s', self.roll_no.get())
        con.commit()
        con.close()   
        self.fetch_data()
        self.clear_data()
        
    def search_data(self):
        con = pymysql.connect(host='localhost', user='root', password='', database='stm')
        cur = con.cursor()
        cur.execute('select * from students where %s LIKE %s', (self.search_by.get().lower(), self.search_txt.get()))
        data = cur.fetchall()
        if len(data) != 0:
            self.st_table.delete(*self.st_table.get_children())
            for row in data:
                self.st_table.insert('', END, values=row)
            con.commit()
        con.close()       
        
        
        
```

```python
root = Tk()
obj = Student(root)
root.mainloop()
```
